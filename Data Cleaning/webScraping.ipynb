```python
import pandas as pd
```

```python
A = []
B = []
for row in my_table.find_all('tr'):
    col = row.find_all('td')
    if(len(col) > 0):
        A.append(col[1].find(text=True))
        B.append(col[5].find(text=True))
df = pd.DataFrame(A, columns=['State'])
df['Population'] = B
```
