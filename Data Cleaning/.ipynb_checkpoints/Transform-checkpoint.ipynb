```python
missing_counts = df.isnull().sum()

missing_counts
```

```python
# Sort variables by the count of missing values in descending order
sorted_missing = missing_counts.sort_values(ascending=False)

# Print the sorted list of variables with missing values
print(sorted_missing)
```
