```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./weather.csv")
```

## Impute Missing Values

```python
rows_with_at_least_one = df[df.isnull().any(axis=1)]
rows_with_at_least_one # rows with at least one missing value
```

```python
rows_missing_values_all_columns = df[df.isnull().all(axis=1)]
rows_missing_values_all_columns # rows which completely empty
```

### Methods to deal with missing values

- **dropna()**: Drop records with missing values.
- **dropna(how='any')**: Drop records if any columns contain a missing value.
- **dropna(how='all')**: Drop records where values in all columns are missing.
- **dropna(axis=1, how='all')**: Drop a column if all the values are missing.
- **dropna(thresh=5)**: Drop records that contain less than 5 non-missing values.

- **fillna(0)**: Replace missing values with zeros.

The function fillna can be used to impute variables / attributes / columns with mean, median, and mode values. 

Example: 

1. mean_val = df[‘column’].mean()

2. df[‘column’].fillna(mean val)

```python
mean_val = df['MinTemp'].mean()
std_val = df['MinTemp'].std()
MinTempZsc = [ (val-mean_val)/std_val for val in df['MinTemp']] # calculate the Z-score
df['MinTempZsc'] = MinTempZsc
```
