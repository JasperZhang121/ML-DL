# Random Forest

Random Forest is an ensemble learning method that combines multiple decision trees to create a more robust and accurate model.
It is based on the concept of bagging (bootstrap aggregating) and uses the idea of an ensemble of decision trees to make predictions, and can be used for both classification and regression tasks.

### Key Features:

- Random Feature Subsets: Each decision tree in the Random Forest is trained on a random subset of features, which helps to decorrelate the trees and increase diversity.

- Bagging: Random Forest uses bagging, which involves creating multiple bootstrap samples from the training data to train individual decision trees.

- Voting: The final prediction in Random Forest is made through majority voting (classification) or averaging (regression) of the predictions from all the individual trees.


## Steps

1.  **Data Preparation**: Prepare the dataset by cleaning, transforming, and encoding the features and labels.
    
2.  **Bootstrap Sampling**: Randomly select a subset of the data (with replacement) to build a decision tree.
    
3.  **Feature Selection**: Randomly select a subset of features to use for each decision tree.
    
4.  **Decision Tree Building**: Build a decision tree using the selected data and features.
    
5.  **Repeat**: Repeat steps 2-4 to build multiple decision trees.
    
6.  **Prediction**: For each test sample, predict the outcome by taking the majority vote of the predictions from all the decision trees.
    
7.  **Evaluate**: Evaluate the performance of the Random Forest model using appropriate metrics such as <mark style="background: #FFB8EBA6;">accuracy, precision, recall, and F1 score</mark>.
    
8.  **Optimize**: Optimize the model by <mark style="background: #ADCCFFA6;">adjusting hyperparameters such as the number of trees, the maximum depth of each tree, and the minimum number of samples required to split a node</mark>.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Select the first two features for simplicity
y = iris.target

# Create a Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X, y)

# Generate a meshgrid of points to create a grid of features
h = 0.02  # Step size for the meshgrid
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, h), np.arange(x2_min, x2_max, h))

# Predict the class labels for each point in the meshgrid
Z = rf_model.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)

# Plot the decision boundaries
plt.figure(figsize=(8, 6))
plt.contourf(xx1, xx2, Z, cmap='coolwarm', alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Random Forest Classifier - Decision Boundaries')
plt.show()
```
